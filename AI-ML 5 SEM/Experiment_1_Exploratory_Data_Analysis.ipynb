# Experiment 1- Exploratory data analysis

### AIM- Exploratory data analysis via statistical descriptions and visualization of iris dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv('iris.csv')
print(df.head())
```

```python
df.Species
```

```python
df.head()
```

```python
df.tail()
```

```python
counts = df.value_counts('Species')
print(counts)
```

```python
species = df.Species.unique()
print(species)
```

```python
df.describe()
```

```python
print(df.std)
```

```python
print(df.std(axis=1))
```

```python
print(df.std(axis=0))
```

```python
print(df.skew(axis=1))
```

```python
print(df.skew(axis=0))
```

```python
df.var(axis=0)
```

```python
print(df.max(axis=0))
```

```python
print(df.min(axis=0))
```

```python
ansArr = [ ]
for i in range(4):
  ansArr.append(df.max(axis=0)[i] - df.min(axis=0)[i])
print(ansArr)
```

```python
sbn.countplot(x='Species', data=df)
```

```python
plt.pie(counts, labels=species, autopct='%1.0f%%')
plt.show()
```

```python
new_df = df.drop(['Id'], axis=1)
new_df.boxplot()
```

```python
sbn.boxplot(x="Species", y="PetalLengthCm", data=df)
plt.show()
```

```python
sbn.violinplot(x="Species", y="PetalLengthCm", data=df)
plt.show()
```

```python
sbn.FacetGrid(df, hue="Species", height=5) \
   .map(sbn.kdeplot, "PetalLengthCm") \
   .add_legend()
plt.show()
```

```python
sbn.pairplot(new_df, hue="Species", height=3)
plt.show()
```
